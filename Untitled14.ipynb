### Function to check if a key exists in a dictionary

There are several ways to check if a key exists in a dictionary. The most Pythonic and common way is to use the `in` operator. This approach is efficient and directly expresses the intent.

```python
def does_key_exist(dictionary, key_to_check):
    """
    Checks if a given key exists in the provided dictionary.

    Args:
        dictionary (dict): The dictionary to search within.
        key_to_check: The key whose existence is to be checked.

    Returns:
        bool: True if the key exists, False otherwise.
    """
    return key_to_check in dictionary

# --- Demonstration ---
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

print(f"Dictionary: {my_dict}")

# Check for an existing key
key1 = "name"
if does_key_exist(my_dict, key1):
    print(f"Key '{key1}' exists in the dictionary.")
else:
    print(f"Key '{key1}' does not exist in the dictionary.")

# Check for a non-existing key
key2 = "country"
if does_key_exist(my_dict, key2):
    print(f"Key '{key2}' exists in the dictionary.")
else:
    print(f"Key '{key2}' does not exist in the dictionary.")

# You can also directly use the 'in' operator without a function for simple checks:
print(f"Is 'age' in my_dict? {'age' in my_dict}")
print(f"Is 'email' in my_dict? {'email' in my_dict}")
```
